```python
'''
Python Libraries for Data Science
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heart.csv")
```

```python
print("First 5 rows:")
print(df.head(), "\n")
```

```python
print("Dataset Info:")
print(df.info(), "\n")
```

```python
possible_targets = ['target', 'output', 'num', 'HeartDisease']
target_col = None
for col in possible_targets:
    if col in df.columns:
        target_col = col
        break

if target_col is None:
    target_col = df.columns[-1]  # use last column if none matches

print(f"Using target column: {target_col}\n")
```

```python
average_age = np.mean(df['age'])
max_chol = np.max(df['chol'])
```

```python
print(f"Average Age: {average_age:.2f}")
print(f"Maximum Cholesterol Level: {max_chol:.2f}\n")
```

```python
plt.hist(df['age'], bins=15, color='lightblue', edgecolor='black')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.countplot(x=target_col, data=df, palette='Set2')
plt.title("Heart Disease Cases (1 = Disease, 0 = No Disease)")
plt.xlabel("Target")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x=target_col, y='chol', data=df, palette='coolwarm')
plt.title("Cholesterol by Heart Disease Status")
plt.xlabel("Target")
plt.ylabel("Cholesterol")
plt.show()
```
